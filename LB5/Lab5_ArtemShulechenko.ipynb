```
def linear_search(a_list, x):

        if a_list[i] == x:
            return i
    return -1

```
def binary_search(a_list, x, left, right):
    if left > right:
        return -1
    mid = (left + right) // 2
    if a_list[mid] == x:
        return mid
    elif a_list[mid] < x:
        return binary_search(a_list, x, mid + 1, right)

````python
import time
import random
import matplotlib.pyplot as plt
import pandas as pd
````

# Бінарний пошук (рекурсивна версія)


````python

def binary_search(arr, x, left, right):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == x:
        return mid
    elif arr[mid] < x:
        return binary_search(arr, x, mid + 1, right)
    else:
        return binary_search(arr, x, left, mid - 1)
````

# Функція для збору часу виконання


````python

def measure_binary_search_times(sizes):
    times = []

    for size in sizes:
        arr = sorted(random.sample(range(size * 10), size))  # відсортований масив
        target = -1  # гарантовано неіснуючий елемент для найгіршого випадку

        start = time.time()
        binary_search(arr, target, 0, len(arr) - 1)
        elapsed = time.time() - start
        times.append(elapsed)

    return times
````

# Розміри масивів


````python

sizes = [10, 100, 500, 1000, 5000, 10000, 50000, 100000, 200000]
````

# Збір даних


````python

binary_times = measure_binary_search_times(sizes)
````

# Таблиця результатів


````python

df = pd.DataFrame({
    "n": sizes,
    "T_binary(n)": binary_times
})
print(df)
````

# Побудова графіка


````python

plt.figure(figsize=(10, 6))
plt.plot(df["n"], df["T_binary(n)"], marker="o", label="Binary Search")
plt.xlabel("Розмір масиву (n)")
plt.ylabel("Час виконання (сек)")
plt.title("Час виконання бінарного пошуку в залежності від розміру масиву")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

# 1. Задача пошуку
Задача пошуку - знайти елемент з певним значенням у структурі даних або встановити його відсутність.
# 2. Критерії оцінки ефективності алгоритмів пошуку

Часова складність
Просторова складність
Простота реалізації
Вимоги до структури даних

# 4. Лінійний пошук
Лінійний пошук - послідовна перевірка кожного елемента до знаходження шуканого.
Асимптотична складність: O(n)
# 5. Бінарний пошук
Бінарний пошук - пошук у відсортованому масиві шляхом ділення інтервалу навпіл.
Принцип: порівнюємо з середнім елементом, відкидаємо половину масиву, повторюємо.
Складність: O(log n)
# 6. Переваги та недоліки бінарного пошуку
Переваги:

Швидкість O(log n)
Передбачуваність

Недоліки:

Потребує відсортованих даних
Складніша реалізація

# 7. Інші алгоритми пошуку

Інтерполяційний пошук: O(log log n), для рівномірних даних
Експоненціальний пошук: O(log n), коли розмір невідомий
Jump Search: O(√n), стрибки + лінійний пошук
Хеш-пошук: O(1) в середньому
Пошук у деревах: O(log n) для збалансованих дерев
