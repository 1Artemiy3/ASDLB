# Вхідні дані 

**Заданий зважений граф: [(1,2,3), (1,3,5), (1,4,7), (2,3,6), (2,4,8), (3,4,9)]**

# Результат роботи алгоритму
* Найкоротший маршрут: (1, 4 )
* Його довжина: 7
# Візуалізація графа

```python
import networkx as nx
import matplotlib.pyplot as plt

# Заданий зважений граф
edges = [(1,2,3), (1,3,5), (1,4,7), (2,3,6), (2,4,8), (3,4,9)]

# Створення графа
G = nx.Graph()
G.add_weighted_edges_from(edges)

# Позиції для вузлів
pos = nx.spring_layout(G, seed=42)

# Візуалізація графа
plt.figure(figsize=(8,6))
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=16)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): w for u, v, w in edges}, font_size=14)
plt.title("Зважений граф", fontsize=16)
plt.show()
```

# Контрольні питання 

# 1. Що таке жадібний алгоритм?
* Жадібний алгоритм — це алгоритм, який на кожному кроці приймає локально оптимальне рішення, сподіваючись, що воно призведе до глобально оптимального результату.
# 2. Які головні принципи роботи жадібних алгоритмів?
* Основні принципи:
* Жадібний вибір: на кожному кроці вибирається найкращий з доступних варіантів (з точки зору деякого критерію).
* Локальна оптимальність: рішення в кожній точці є найкращим на даний момент.
* Незмінність підзадачі: після прийняття рішення, залишок задачі має бути подібним до початкової, тобто можна застосовувати той самий підхід.
# 3. Яка головна відмінність між жадібними алгоритмами та динамічним програмуванням?
* Критерій	Жадібні алгоритми	Динамічне програмування
* Підхід	Локально оптимальні вибори	Повний перебір усіх підзадач
* Гарантія оптимальності	Не завжди	Завжди (для правильно сформульованих задач)
* Повторне використання рішень	Немає	Є (мемоізація/таблиці)
* Складність реалізації	Проста	Складніша
# 4. Приклади задач, які можна розв’язати за допомогою жадібних алгоритмів:
* Задача про монети (мінімальна кількість монет для суми) – тільки для деяких валютних систем.
* Задача Кангуру і дерево (макс. покриття вузлів).
* Задача про рюкзак (лише дробова версія — fractional knapsack).
* Алгоритм Крускала або Прима для пошуку мінімального остовного дерева (MST).
* Алгоритм Дейкстри — знаходження найкоротших шляхів.
* Планування розкладу: задача про інтервали (interval scheduling).
# 5. Які можуть бути обмеження у використанні жадібних алгоритмів для розв'язання екстремальних задач?
* Немає гарантії глобальної оптимальності: жадібне рішення не завжди веде до найкращого результату.
* Залежність від задачі: працює лише в задачах, де дотримується властивість жадібної оптимальності.
* Не можна відкотитись назад: немає повернення до попередніх рішень, навіть якщо вони були помилковими.
* Часто потребує доказу правильності, що складно для складних задач.
# 6. Чому жадібні алгоритми часто використовуються для наближеного розв’язання екстремальних задач?
* Швидкість: Жадібні алгоритми працюють значно швидше, ніж повні перебори або динамічне програмування.
* Простота реалізації: менше коду, менше ресурсів.
* Дають хороші наближення: для багатьох NP-складних задач жадібні стратегії дають розв’язки, близькі до оптимальних (наприклад, 
* задачі про покриття множини або рюкзак).
* Часто використовуються в евристиках, жадібних апроксимаційних алгоритмах та онлайн-алгоритмах.
