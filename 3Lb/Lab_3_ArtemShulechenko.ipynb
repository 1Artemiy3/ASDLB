```python
import time
import random
import matplotlib.pyplot as plt
import pandas as pd

# Алгоритми сортування
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Розширений список розмірів масиву
sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]
insert_times, bubble_times = [], []

for size in sizes:
    data = [random.randint(0, 1000000) for _ in range(size)]

    # Тест сортування вставками (лише до 2000, бо повільно)
    if size <= 2000:
        start = time.time()
        insertion_sort(data.copy())
        insert_times.append(time.time() - start)
    else:
        insert_times.append(None)  # Пропускаємо, надто довго

    # Тест сортування бульбашкою (лише до 2000)
    if size <= 2000:
        start = time.time()
        bubble_sort(data.copy())
        bubble_times.append(time.time() - start)
    else:
        bubble_times.append(None)

# Створення таблиці
df = pd.DataFrame({
    "n": sizes,
    "T_insert(n)": insert_times,
    "T_bubble(n)": bubble_times
})

# Виведення таблиці
print(df)

# Побудова графіка (тільки не-None значення)
plt.figure(figsize=(12, 6))
plt.plot(df["n"], df["T_insert(n)"], label="Insertion Sort", marker="o")
plt.plot(df["n"], df["T_bubble(n)"], label="Bubble Sort", marker="s")
plt.xlabel("Розмір масиву (n)")
plt.ylabel("Час виконання (сек)")
plt.title("Залежність часу T(n) для insertion та bubble сортування")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 1. O-нотація та Ω-нотація
O-нотація (велике O) описує верхню межу часової складності алгоритму - найгірший можливий випадок. Вона показує, що час виконання не перевищуватиме певної функції при достатньо великих значеннях вхідних даних.
Ω-нотація (велика омега) описує нижню межу часової складності - найкращий можливий випадок. Вона показує, що час виконання не буде меншим за певну функцію.
Основна відмінність: O-нотація обмежує зверху (максимум), а Ω-нотація обмежує знизу (мінімум).
# 2. Сортування вставляннями для відсортованого масиву
Часова складність: O(n)
Обґрунтування:

У найкращому випадку (коли масив уже відсортований) алгоритм проходить масив лише один раз
Для кожного елемента виконується лише одне порівняння з попереднім елементом
Оскільки умова порівняння завжди хибна (поточний елемент уже на своєму місці), внутрішній цикл не виконується
Загальна кількість операцій пропорційна n, тому складність O(n)

# 3. Ефективний алгоритм
Ефективний алгоритм - це алгоритм, який розв'язує задачу за поліноміальний час відносно розміру вхідних даних. Такі алгоритми мають часову складність O(n^k), де k - константа, а n - розмір входу. Ефективні алгоритми протиставляються експоненціальним алгоритмам O(2^n), які вважаються неефективними для великих обсягів даних.
# 4. Головний параметр (розмір) завдання
Головний параметр - це характеристика вхідних даних, яка найбільше впливає на складність алгоритму. Це може бути:

Кількість елементів у масиві (для сортування)
Кількість вершин у графі (для алгоритмів на графах)
Розрядність числа (для арифметичних операцій)
Довжина рядка (для роботи з текстом)

Вибір правильного головного параметра критично важливий для коректного аналізу складності.
# 5. Асимптотична складність для F(N) = N³ + 7N² - 14N
Відповідь: O(N³)
Пояснення:
В O-нотації враховується лише найстарший член поліному (з найвищим степенем), оскільки при великих значеннях N він домінує над іншими членами. Коефіцієнти також опускаються, тому що вони не впливають на порядок зростання функції.
N³ зростає значно швидше ніж 7N² або 14N при великих N, тому асимптотична складність дорівнює O(N³).
